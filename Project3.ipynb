```r
install.packages("car")
```

```r
#example
wrat_r = c(115,98,107,90,85,90,100,105,95,70,85,78,89,100,90,65,80,72)
group = c(rep("Treatment",3),rep("Control",3))
level = c(rep("Mild",6),rep("Moderate",6),rep("Severe",6))
mdataframe = data.frame(wrat_r,group,level)
mdataframe
```

```r
# Check Normality by Shapiro
with(mdataframe,shapiro.test(wrat_r[level == '1']))
```

```r
# Check variance test before ANOVA
library(car)
leveneTest(wrat_r ~ group * level, data = mdataframe, center = mean)
```

```r
# 1. สร้าง Model Two-way ANOVA
model2way <- aov(wrat_r ~ group * level , data = mdataframe)

# 2. ดึงค่า Residuals ออกมาทดสอบ Shapiro-test ครั้งเดียวจบ
shapiro.test(residuals(model2way))
```

```r
# Check ANOVA
aov.model = aov(wrat_r ~ group * level , data = mdataframe)
summary(aov.model)
```

```r
# Check Tukey Honest Significant Difference
TukeyHSD(aov.model)
```
